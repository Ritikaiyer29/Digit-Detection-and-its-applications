```python
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import cv2
import pytesseract

# Function to download and preprocess the image
def download_image(image_url):
    try:
        response = requests.get(image_url, timeout=10)
        response.raise_for_status()  # Raise an error for bad responses
        img = Image.open(BytesIO(response.content))
        img.verify()  # Verify that the image is not corrupted
        return img
    except Exception as e:
        print(f"Error downloading or verifying image from {image_url}: {e}")
        return None

# Function to convert image to OpenCV format and preprocess
def preprocess_image(image_url):
    img = download_image(image_url)
    if img is None:
        return None  # Return None if the image couldn't be downloaded or is corrupted

    # Convert the PIL image to OpenCV format
    img = np.array(img)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  # Convert from RGB to BGR

    # Convert the image to grayscale and apply threshold
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    return thresh

# Function to extract text from the processed image
def extract_text(image_url):
    processed_image = preprocess_image(image_url)
    if processed_image is None:
        return ""  # Return empty string if the image couldn't be processed

    text = pytesseract.image_to_string(processed_image)
    return text

# Example DataFrame
import pandas as pd
data = pd.DataFrame({
    'index': [1, 2],
    'image_link': ['https://example.com/image1.jpg', 'https://example.com/image2.jpg']
})

# Applying the function to your dataset
data['ocr_text'] = data['image_link'].apply(extract_text)

# Check the extracted text
data.head()
```
